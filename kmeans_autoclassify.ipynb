```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
pca_train = pd.read_csv('data/train_pca2comp.csv',index_col = 0)
pca_test = pd.read_csv('data/test_pca2comp.csv',index_col = 0)
base = pd.read_csv("data/train_numeric.csv", usecols = ['Id','Response'])
```

```python
kmean = KMeans(n_clusters=4)
fit = kmean.fit(pca_train)
```

```python
base['labels'] = fit.labels_
```

```python
print(base.groupby('labels').Response.count(),
      base.groupby('labels').Response.sum()/base.groupby('labels').Response.count())
```
